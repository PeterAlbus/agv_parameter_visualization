```python
import matplotlib.pyplot as plt
import numpy as np

v_creep = np.float64(0.12)


def plot_condition(
    *,
    label: str,
    colors: tuple[str, str],
    v: np.ndarray,
    a: np.float64,
    d_creep: np.float64,
) -> None:
    s_0 = 0.5 * (v ** 2 - v_creep ** 2) / a
    s_acc  = s_0 + 1.6 * d_creep + 0.5 * v
    s_dec = s_acc + 1.0 * v + 1.0
    plt.plot(v, s_acc, "--", color=colors[0], label=f"s_acc ({label})")
    plt.plot(v, s_dec, "--", color=colors[1], label=f"s_dec ({label})")


plt.figure(figsize=(12, 8))

plot_condition(
   label="empty",
   colors=("cyan", "blue"),
   v=np.linspace(v_creep, 6.0, 1001),
   a=np.float64(6 / 9),
   d_creep=np.float64(0.2),
)

plot_condition(
    label="loaded",
    colors=("orange", "red"),
    v=np.linspace(v_creep, 3.0, 1001),
    a=np.float64(3 / 8),
    d_creep=np.float64(0.3),
)

v = np.linspace(v_creep, 6.0, 1001)
d = 10 * v + 2
plt.plot(v, d, "-", color="k", label="VMS Length")
plt.plot(v, d - 1 * v, "-", color="gray", label="VMS Length (after 1s)")

plt.gca().set(
    ylabel="Distance (m)",
    xlabel="Speed (m/s)",
)
plt.legend()
plt.grid(alpha=0.2)

pass
```
